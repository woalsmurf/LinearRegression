```python
# Import Modules

import numpy as np
import matplotlib.pyplot as plt
import random
from datetime import datetime
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Gain Permission to Google Drive

from google.colab import drive
drive.mount('/content/drive')

# Declare LinearRegression Object

regressor_lin = LinearRegression()                                                                                                                                                          

# Read Data

file_input = open("/content/drive/MyDrive/Colab_Notebooks/LinearRegression_30835/input.txt", "r")
file_target = open("/content/drive/MyDrive/Colab_Notebooks/LinearRegression_30835/target.txt", "r")

list_input = file_input.readlines()
list_target = file_target.readlines()

# Remove Space and \n

for i in range(len(list_input)):
  list_input[i] = list_input[i].strip(' \n')

for i in range(len(list_target)):
  list_target[i] = list_target[i].strip(' \n')

# Extract Names from Data

name_input = list_input[0]
name_target = list_target[0]

del list_input[0]
del list_target[0]

# Compare the Length of Input and Target

if len(list_input) != len(list_target):
  raise Exception("input.txt와 target.txt의 데이터 수가 각각 {}개와 {}개로, 서로 다릅니다.".format(len(list_input), len(list_target)))

length_list = len(list_input)

# Convert Input to float

list_error = []

for i in range(length_list):
  try:
    list_input[i] = float(list_input[i])
  except Exception:
    list_error.append(i+2)

if list_error == []:
  print("# {} 데이터를 성공적으로 변환했습니다.".format(name_input))
  print(list_input, end="\n\n")
else:
  raise Exception("input.txt의 {} 행에서 데이터 변환에 실패했습니다.".format(list_error))

# Convert Target to float

list_error = []

for i in range(length_list):
  try:
    list_target[i] = float(list_target[i])
  except Exception:
    list_error.append(i+2)

if list_error == []:
  print("# {} 데이터를 성공적으로 변환했습니다.".format(name_target))
  print(list_target, end="\n\n")
else:
  raise Exception("target.txt의 {} 행에서 데이터 변환에 실패했습니다.".format(list_error))

# Separate Sets

random.seed(datetime.now())

list_train_input, list_test_input, list_train_target, list_test_target = train_test_split(list_input, list_target, random_state=random.randint(0, 2*32-1))

length_train = len(list_train_input)
length_test = len(list_test_input)

print("# 분리한 train_input({}): {}".format(length_train, list_train_input))
print("# 분리한 test_input({}): {}".format(length_test, list_test_input))
print("# 분리한 train_target({}): {}".format(length_train, list_train_target))
print("# 분리한 test_target({}): {}".format(length_test, list_test_target))
print()

# Show Sets

plt.scatter(list_train_input, list_train_target)
plt.scatter(list_test_input, list_test_target)
plt.xlabel(name_input)
plt.ylabel(name_target)
plt.show()

# Convert List to Array

train_input = np.array(list_train_input)
test_input = np.array(list_test_input)
train_target = np.array(list_train_target)
test_target = np.array(list_test_target)

# Reshape Input Sets

train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

train_current = train_input
test_current = test_input

# Learn and Predict

print("# 훈련과 테스팅, 예측을 시작합니다.")

while True:

  # Get Degree

  print()
  print("# 다항 회귀 모델의 차수를 입력하십시오. ")
  degree = input("# 0을 입력하면 종료합니다. -1을 입력하면 세트를 재조합합니다. ")

  if int(degree) < -1:
    print("# 차수는 1보다 크거나 같아야 합니다. 다시 입력하십시오.")
    continue

  elif int(degree) == 0:
    print("# 프로그램을 종료합니다.")
    break

  elif int(degree) == -1:

    # Same Process before While Repeating

    # Separate Sets

    random.seed(datetime.now())

    list_train_input, list_test_input, list_train_target, list_test_target = train_test_split(list_input, list_target, random_state=random.randint(0, 2*32-1))

    length_train = len(list_train_input)
    length_test = len(list_test_input)

    print("# 분리한 train_input({}): {}".format(length_train, list_train_input))
    print("# 분리한 test_input({}): {}".format(length_test, list_test_input))
    print("# 분리한 train_target({}): {}".format(length_train, list_train_target))
    print("# 분리한 test_target({}): {}".format(length_test, list_test_target))
    print()

    # Show Sets

    plt.scatter(list_train_input, list_train_target)
    plt.scatter(list_test_input, list_test_target)
    plt.xlabel(name_input)
    plt.ylabel(name_target)
    plt.show()

    # Convert List to Array

    train_input = np.array(list_train_input)
    test_input = np.array(list_test_input)
    train_target = np.array(list_train_target)
    test_target = np.array(list_test_target)

    # Reshape Input Sets

    train_input = train_input.reshape(-1, 1)
    test_input = test_input.reshape(-1, 1)

    continue

  elif not degree.isdigit():
    print("# 정수로 변환 가능한 데이터가 아닙니다. 다시 입력하십시오.")
    continue

  else:
    degree = int(degree)
  
  # Reshape Data by Degree
  train_current = train_input
  test_current = test_input

  if degree != 1:
    for i in range(2, degree+1):
      train_current = np.column_stack((train_input ** i, train_current))
      test_current = np.column_stack((test_input ** i, test_current))

  # Train and Test LinearRegressor

  regressor_lin.fit(train_current, train_target)
  coef = regressor_lin.coef_
  intercept = regressor_lin.intercept_

  print()
  print("# 훈련을 완료했습니다.")
  print("# 선형 회귀를 통해 얻은 다항식의 계수: {}".format(coef))
  print("# 선형 회귀를 통해 얻은 다항식의 y절편: {}".format(intercept))

  train_score = regressor_lin.score(train_current, train_target)
  test_score = regressor_lin.score(test_current, test_target)

  print()
  print("# 테스트를 완료했습니다.")
  print("# 훈련 세트의 테스트 스코어: {}".format(train_score))
  print("# 테스트 세트의 테스트 스코어: {}".format(test_score))

  print()
  try:
    predict = int(input("아무 수를 입력하면 예측을 시작합니다. 0을 입력하면 모델을 재훈련합니다. "))
  except:
    predict = 0

  while predict:
    
    # Get User Input

    print()
    print("{}을 예측할 {} 값을 입력하세요.".format(name_target, name_input))
    user_input_num = input("수가 아닌 값을 입력하면 예측을 종료합니다. ")

    try:
      user_input = np.append([], float(user_input_num))
    except:
      print("예측을 종료합니다.")
      break

    # Reshape User Input by Degree

    user_degree = user_input
    if degree != 1:
      for i in range(2, degree+1):
        user_degree = np.column_stack((user_input ** i, user_degree))

    # Predict User Target

    user_target = regressor_lin.predict(user_degree)

    print()
    print("예측한 {} = {} 의 {}: {}".format(name_input, user_input[0], name_target, user_target[0]))

    # Show Prediction

    plt.scatter(list_train_input, list_train_target)
    plt.scatter(list_test_input, list_test_target)
    plt.scatter(user_input, user_target, marker='^', color='red')
    plt.xlabel(name_input)
    plt.ylabel(name_target)
    plt.show()
```
